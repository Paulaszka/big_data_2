```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Część 5
### Misja dodatkowa - Korelacja danych

##### Współczynnik korelacji przedstawia związek pomiędzy dwoma zmiennymi, co pozwala na zaobserwowanie potencjalnych zależności pomiędzy informacjami. 

##### Stwórz macierze korelacji, które przedstawią współczynniki korelacji dla każdej pary zmiennych:


##### 5.1. liczba nowych zachorowań, szczepień i zgonów, a liczba nowych zachorowań, szczepień i zgonów,

```python
new_metrics_df = df[['new_confirmed', 'new_persons_vaccinated', 'new_deceased']]
new_metrics_corr = new_metrics_df.corr()

plt.figure(figsize=(10, 4))
sns.heatmap(new_metrics_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 5.2. liczba nowych zachorowań, szczepień, zgonów, a łączna liczba zachorowań, szczepień, zgonów,

```python
cumulative_metrics_df = df[['cumulative_confirmed', 'cumulative_persons_vaccinated', 'cumulative_deceased']]
cumulative_metrics_corr = cumulative_metrics_df.corr()

plt.figure(figsize=(10, 4))
sns.heatmap(cumulative_metrics_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 5.3. wskaźniki gospodarcze, a liczba nowych zachorowań, szczepień i zgonów,

```python
metrics_df = df[['gdp_usd', 'new_confirmed', 'new_persons_vaccinated', 'new_deceased']]
metrics_corr = metrics_df.corr()

plt.figure(figsize=(10, 4))
sns.heatmap(metrics_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 5.4. parametry demograficzne, a liczba nowych zachorowań, szczepień i zgonów,

```python
demographic_columns = ['population', 'area_sq_km', 'density_sq_km', 'growth_rate', 'world_population_percentage']
new_metrics_columns = ['new_confirmed', 'new_persons_vaccinated', 'new_deceased']
demographic_df = df[demographic_columns]

combined_df = pd.concat([demographic_df, new_metrics_df], axis=1)
full_corr = combined_df.corr()
cross_corr = full_corr.loc[demographic_columns, new_metrics_columns]

plt.figure(figsize=(10, 4))
sns.heatmap(cross_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

##### 5.5. co najmniej 4 dodatkowe korelacje pomiędzy różnymi danymi.

```python
base_columns = ['cumulative_confirmed', 'cumulative_persons_vaccinated', 'cumulative_deceased', 'cumulative_tested']
```

```python
gdp_growth_columns = ['gdp_in_1995', 'gdp_in_2000', 'gdp_in_2005', 'gdp_in_2010', 'gdp_in_2015', 'gdp_in_2016', 'gdp_usd']

gdp_df = df[gdp_growth_columns]
covid_outcomes_df = df[base_columns]

combined_gdp_covid = pd.concat([gdp_df, covid_outcomes_df], axis=1)
gdp_covid_corr = combined_gdp_covid.corr().loc[gdp_growth_columns, base_columns]

plt.figure(figsize=(12, 6))
sns.heatmap(gdp_covid_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation: GDP Growth Metrics by by New Confirmed, Vaccinated, Deceased and Tested')
plt.xlabel('New Confirmed, Vaccinated, Deceased and Tested')
plt.ylabel('GDP Metrics')
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
population_columns = ['1970_population', '1980_population', '1990_population', '2000_population', 
                      '2010_population', '2015_population', '2020_population', '2022_population']

pop_test_df = pd.concat([df[population_columns], df[base_columns]], axis=1)
pop_test_corr = pop_test_df.corr().loc[population_columns, base_columns]

plt.figure(figsize=(12, 6))
sns.heatmap(pop_test_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation: Population Changes Over Time by New Confirmed, Vaccinated, Deceased and Tested')
plt.xlabel('COVID Testing and Cases')
plt.ylabel('Population by Year')
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
geographic_df = pd.get_dummies(df[['continent']], prefix=['continent'])
geo_covid_df = pd.concat([geographic_df, df[base_columns]], axis=1)
geo_covid_corr = geo_covid_df.corr().loc[geographic_df.columns, base_columns]

plt.figure(figsize=(12, 6))
sns.heatmap(geo_covid_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation: Continents by New Confirmed, Vaccinated, Deceased and Tested')
plt.xlabel('New Confirmed, Vaccinated, Deceased and Tested')
plt.ylabel('Continents')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
demographic_columns = ['population', 'area_sq_km', 'density_sq_km', 'growth_rate', 'world_population_percentage']
health_metrics_per_1000 = ['confirmed_per_1000', 'vaccinated_per_1000', 'deceased_per_1000']

demographic_df = df[demographic_columns]
health_metrics_df = df[health_metrics_per_1000]

combined_df = pd.concat([demographic_df, health_metrics_df], axis=1)
full_corr = combined_df.corr()
cross_corr = full_corr.loc[demographic_columns, health_metrics_per_1000]

plt.figure(figsize=(10, 4))
sns.heatmap(cross_corr, annot=True, vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix: Demographic Parameters by New Confirmed, Vaccinated and Deceased')
plt.xlabel('Health Metrics')
plt.ylabel('Demographic Parameters')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
